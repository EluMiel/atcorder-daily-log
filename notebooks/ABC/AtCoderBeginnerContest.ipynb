### ABC081B - Shift only

### ABC083B - Some Sums

```python
n, a, b = map(int, input().split())
ans = 0  # 総和を代入する変数

for x in range(1, n+1):  # 1以上N以下の数を取り出す
    num = list(str(x))  # xをstr型で各桁に分解
    digit_sum = sum(list(map(int, num)))  # 各桁の総和を計算
    if a <= digit_sum <= b:
        ans += x

print(ans)
```

- 総和の取り方はワンライナーにできる。 digit_sum = sum(map(int, str(x)))
- for文のコメントは「1からNまでの整数を順に処理」の方が自然。
- 下一桁でバッサリ枝狩りは理屈では可能なものの、効果の割にコードがかなり複雑化するので不要

### ABC085C - Otoshidama(TLE)

```python
n, y = map(int, input().split())  # n:お札の枚数 y:お年玉の合計金額

def check_otoshidama(n, y):  #お年玉の金額と枚数について検証する関数。
    y2 = y // 1000  #お札の金額、お年玉の合計金額は1000で割った数で計算する。
    if n > y2 or y2 > 10 * n:  #お札の枚数が合計金額/1000よりも多いときは全て1000円でもオーバーするので嘘に決定。
        return (-1, -1, -1)
    else:
        for z in range(n + 1):  #1000円の枚数
            if z == y2 and z == n:
                return (0, 0, z)
            for y in range(n - z + 1):  #5000円の枚数
                if (5 * y) + z == y2 and y + z == n:
                    return (0, y, z)
                for x in range(n - y - z + 1):  #10000円の枚数
                    if (10 * x) + (5 * y) + z == y2 and x + y + z == n:
                        return (x, y, z)
        return (-1, -1, -1)

print(*check_otoshidama(n, y))
```

- 合計金額と5000円の枚数両方に変数yを使っているのは可読性的にアウト。誤解を招く。Pythonの動きとしては、2つめのyが現れた時点で1つめのyは上書きされるので問題ない。
- y2 = y / 1000 → y2 = y // 1000 小数を避けるために整数除算にすること。
- n > y2 or y2 > 10*n → y2 < n or y2 > 10n 意味は同じでも右の方が見やすい。
- print内は * (アンパック演算子)を使って記述しないとAtCorder指定の出力にならない。

### ABC049C - 白昼夢

```python
words = input()

while words:  # wordsが空になるまでdream dreamer erase eraserを取り出す。
    if words[0:7] == "dreamer" and words[5:10] != "erase":  # 末尾のerがeraseの一部の場合はdreamとして取り出す。
        words = words[7:]
    elif words[0:6] == "eraser":
        words = words[6:]
    elif words[0:5] == "dream":
        words = words[5:]
    elif words[0:5] == "erase":
        words = words[5:]
    else:
        break

if words:  # wordsにdream dreamer erase eraser以外の文字列が残っている場合。
    print("NO")
else:
    print("YES")
```

- 前方貪欲+特例、スライスでの新規文字列生成を用いたコード。
- wordsは空文字列になるとFalseを返すのでWhile文の判定条件として使う。
- Pythonのスライスははみ出しても例外にならないので境界アクセスが安全。
- **前方貪欲+特例は壊れやすい。**  
パターンが増えるとifが増殖しがちで保守性が低下。特例の作り方をミスすると一発でエラーコードになる危険性。
- **スライスで毎回コピーの問題点。**  
新しい文字列を都度生成するのはオーダー増加の原因になる。(O(n**2)に寄る。)
- **後方探索+ポインタ方式が安全でメモリ効率が良い。**

### ABC049C - 白昼夢(後方探索+ポインタ方式)

```python
words = input()
i = len(words)
while i > 0:
    if i >= 7 and words.endwith("dreamer", 0, i): i -= 7
    elif i >= 6 and words.endwith("eraser", 0, i): i -= 6
    elif i >= 5 and words.endwith("dream", 0, i) or words.endwith("erase", 0, i): i -= 5
    else:
        print("NO"); break
else:
    print("YES")
```

- endwith(suffix, start=0, end=len(string))
- Pythonではif/ for/ while どれでも : の後に1つの文だけ書ける。
- セミコロン(;)を使うと2文を1行にまとめられるがPEP8では非推奨。
- for while における else はループが"**breakされずに最後まで回り切ったとき**"に実行される。
- 条件 i >= 7 は無駄な比較を発生させないためのもの。

### ABC086C - Traveling

```python
n = int(input())
list_t = [0]
list_x = [0]
list_y = [0]
cnt = 0  # AtCoDeerくんの移動成功回数

for _ in range(n):
    t, x, y = map(int, input().split())
    list_t.append(t)
    list_x.append(x)
    list_y.append(y)

for i in range(1, n + 1):
    time = list_t[i] - list_t[i - 1]  # 前回の点から今回の点への移動にかけることができる時間
    dist = abs(list_x[i] - list_x[i - 1]) +  abs(list_y[i] - list_y[i - 1])  # 前回の点から今回の点への移動距離(時間)
    if time == dist:
        cnt += 1
    elif time > dist:
        if (time - dist) % 2 == 0:  # 移動時間が余る場合は余分時間が偶数でないと目的地に戻れない。
            cnt += 1
        else:
            break
    elif time < dist:
        break

if cnt == n:
    print("Yes")
else:
    print("No")
```

- 修正点はメモリの削減と、早期breakの整理
- 値をすべてリストに保存するとO(n) → 前回値だけもてばO(1)
- breakはifが何段ネストしていても**一番近いループを抜ける**ことができる。


### ABC086C - Traveling(模範解答/区間ごとに検査して早期リターンのテンプレ)

```python
n = int(input())
prev_t, prev_x, prev_y = 0, 0, 0

for _ in range(n):
    t, x, y = map(int, input().split())
    dt = t - prev_t
    dist = abs(x - prev_x) + abs(y - prev_y)
    if dt < dist or (dt - dist) % 2 != 0:
        print("No")
        break
    prev_t, prev_x, prev_y = t, x, y
else:
    print("Yes")
```

- 変数prevで前回の値をキープする
- 条件をワンライナーで表現する
- for文を完走で抜けたのか、breakで抜けたのかの判断をfor-elseで行う。(while/forのelseは"**breakせずに完走したとき**"にのみ動作する。)

C - Robot Factory

```python
n, m, k = map(int, input().split())
list_h = [int(h) for h in input().split()]
list_b = [int(b) for b in input().split()]
list_h.sort()
list_b.sort()
stop = True
cnt = 0

while stop:
  for i in range(n):
    if not stop:
      break
    for j in range(m - cnt):
        if list_h[i] <= list_b[j]:
            list_b.pop(j)
            cnt += 1
            break
        if cnt == k:
            stop = False
            break
  else:
    stop = False

print("Yes" if cnt == k else "No")
```
